```python
# Identificar outliers con IQR en horas_estudio_semanal y monto_inversion
def detectar_outliers(col):
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_clean[(df_clean[col] < lower_bound) | (df_clean[col] > upper_bound)]
    return outliers[[col]]

print("\nOutliers en horas_estudio_semanal:")
print(detectar_outliers("horas_estudio_semanal"))

print("\nOutliers en monto_inversion:")
print(detectar_outliers("monto_inversion"))
```

```python
import pandas as pd
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder, FunctionTransformer

# Cargar dataset
df = pd.read_csv("Estudiantes_edutech.csv")

# Creacion nuevas variables

def crear_variables_nuevas(X):
    X = X.copy()
    X['tasa_completitud'] = (X['tareas_completadas'] / X['tareas_totales']) * 100
    X['estudiante_activo'] = np.where(X['sesiones_totales'] > 50, 1, 0)
    # Rellenar posibles divisiones por cero o NaNs
    X['tasa_completitud'] = X['tasa_completitud'].fillna(0)
    return X

# Codificación de variables categóricas

nominal_vars = ['genero', 'pais', 'dispositivo_principal', 'horario_estudio', 'plataforma_origen', 'experiencia_previa']
numeric_vars = ['edad', 'horas_estudio_semanal', 'sesiones_totales', 'tareas_completadas',
                'participacion_foros', 'promedio_evaluaciones', 'monto_inversion',
                'estudiante_activo', 'tasa_completitud']
ordinal_vars = ['nivel_educacion']

# Crear el Preprocesador (ColumnTransformer)
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_vars),
        ('cat_nom', OneHotEncoder(drop='first', sparse_output=False), nominal_vars),
        ('cat_ord', OrdinalEncoder(categories=[['Secundaria', 'Pregrado', 'Posgrado']]), ordinal_vars)
    ],
    remainder='drop'
)

# Construir el Pipeline Final
pipeline = Pipeline(steps=[
    ('feature_eng', FunctionTransformer(crear_variables_nuevas, validate=False)),
    ('preprocessor', preprocessor)
])

# Uso:
df = pd.read_csv('Estudiantes_edutech.csv')
X_procesado = pipeline.fit_transform(df)
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Separar features y variables objetivo
X = df.drop(columns=['curso_completado', 'puntaje_satisfaccion'])
y_clas = df['curso_completado']
y_reg = df['puntaje_satisfaccion']

# Dividir en entrenamiento (70%) y prueba (30%)
# Para clasificación (usamos stratify para mantener proporciones)
X_train, X_test, y_train_clas, y_test_clas = train_test_split(
    X, y_clas, test_size=0.30, random_state=42, stratify=y_clas
)

# Para regresión
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_reg, test_size=0.30, random_state=42
)

# Procesar los datos con el pipeline antes de aplicar SMOTE
# SMOTE necesita datos numéricos, por lo que primero aplicamos fit y transformamos
X_train_prepadado = pipeline.fit_transform(X_train)
X_test_preparado = pipeline.transform(X_test)

# Aplicar SMOTE solo al set de entrenamiento de clasificación
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_prepadado, y_train_clas)

print(f"Dataset original: {y_train_clas.value_counts().to_dict()}")
print(f"Dataset con SMOTE: {pd.Series(y_train_balanced).value_counts().to_dict()}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, roc_auc_score, confusion_matrix, classification_report)
import seaborn as sns
import matplotlib.pyplot as plt

# Definir modelos y sus mallas de hiperparámetros
models_data = {
    'RandomForest': {
        'model': RandomForestClassifier(random_state=42),
        'params': {
            'n_estimators': [100, 200],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5]
        }
    },
    'LogisticRegression': {
        'model': LogisticRegression(max_iter=1000, solver='liblinear', random_state=42),
        'params': {
            'C': [0.1, 1, 10],
            'penalty': ['l1', 'l2']
        }
    }
}

best_models = {}

# Entrenamiento con GridSearchCV
for name, config in models_data.items():
    print(f"Optimizando {name}...")
    grid = GridSearchCV(config['model'], config['params'], cv=5, scoring='f1', n_jobs=-1)
    grid.fit(X_train_balanced, y_train_balanced)
    best_models[name] = grid.best_estimator_
    print(f"Mejores parámetros {name}: {grid.best_params_}\n")

# Evaluación y Comparación
for name, model in best_models.items():
    y_pred = model.predict(X_test_preparado)
    y_proba = model.predict_proba(X_test_preparado)[:, 1]

    print(f"--- Reporte de Evaluación: {name} ---")
    print(f"Accuracy: {accuracy_score(y_test_clas, y_pred):.4f}")
    print(f"ROC-AUC: {roc_auc_score(y_test_clas, y_proba):.4f}")
    print(classification_report(y_test_clas, y_pred))

    # Matriz de Confusión
    cm = confusion_matrix(y_test_clas, y_pred)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Matriz de Confusión - {name}")
    plt.xlabel("Predicho")
    plt.ylabel("Real")
    plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns

# Definir modelos y mallas para Regresión
reg_models_data = {
    'RandomForestRegressor': {
        'model': RandomForestRegressor(random_state=42),
        'params': {
            'n_estimators': [50, 100],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5]
        }
    },
    'Ridge': {
        'model': Ridge(),
        'params': {
            'alpha': [0.1, 1.0, 10.0, 100.0]
        }
    }
}

best_reg_models = {}

# GridSearchCV
for name, config in reg_models_data.items():
    grid = GridSearchCV(config['model'], config['params'], cv=5, scoring='neg_mean_absolute_error', n_jobs=-1)
    grid.fit(X_train_prepadado, y_train_reg)
    best_reg_models[name] = grid.best_estimator_
    print(f"Mejores parámetros {name}: {grid.best_params_}")

# Evaluación y Gráfico (usando el mejor modelo, usualmente RandomForest)
for name, model in best_reg_models.items():
    y_pred = model.predict(X_test_preparado)

    print(f"\n--- Métricas {name} ---")
    print(f"R²: {r2_score(y_test_reg, y_pred):.4f}")
    print(f"MAE: {mean_absolute_error(y_test_reg, y_pred):.4f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_test_reg, y_pred)):.4f}")

    # Gráfico Real vs Predicho
    plt.figure(figsize=(6,4))
    sns.scatterplot(x=y_test_reg, y=y_pred, alpha=0.5)
    plt.plot([y_test_reg.min(), y_test_reg.max()], [y_test_reg.min(), y_test_reg.max()], '--r')
    plt.title(f"Reales vs Predicciones - {name}")
    plt.xlabel("Valores Reales")
    plt.ylabel("Predicciones")
    plt.show()

# Identificar las 5 variables más importantes (basado en RandomForest)
rf_final = best_reg_models['RandomForestRegressor']
importancias = rf_final.feature_importances_

# Recuperar nombres de columnas del pipeline (se asume el orden del preprocessor)
features = numeric_vars + \
           list(pipeline.named_steps['preprocessor'].transformers_[1][1].get_feature_names_out(nominal_vars)) + \
           ordinal_vars

feat_imp = pd.Series(importancias, index=features).sort_values(ascending=False)

print("\n--- Top 5 Variables más importantes ---")
print(feat_imp.head(5))

# Gráfico de importancia
plt.figure(figsize=(8,4))
feat_imp.head(5).plot(kind='barh', color='teal').invert_yaxis()
plt.title("Top 5 Importancia de Features")
plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import when, col
from pyspark.ml.feature import StringIndexer, VectorAssembler, StandardScaler
from pyspark.ml import Pipeline

# Crear la Sesión de Spark
spark = SparkSession.builder.appName("EdutechAnalysis").getOrCreate()

# Cargar el dataset
# Nota: Asegúrate de que el archivo CSV esté accesible en la ruta
df_spark = spark.read.csv("Estudiantes_edutech.csv", header=True, inferSchema=True)

# Crear la columna label binaria (curso_completado)
df_spark = df_spark.withColumn("label", col("curso_completado").cast("double"))

# Definir las variables
nominal_vars = ['genero', 'pais', 'dispositivo_principal', 'horario_estudio', 'plataforma_origen', 'nivel_educacion']
numeric_vars = ['edad', 'horas_estudio_semanal', 'sesiones_totales', 'tareas_completadas',
                'participacion_foros', 'promedio_evaluaciones', 'monto_inversion']

# Preparar los Transformadores
# StringIndexer: Convierte categorías (texto) a índices numéricos
indexers = [StringIndexer(inputCol=c, outputCol=f"{c}_indexed", handleInvalid="keep")
            for c in nominal_vars]

# VectorAssembler: Junta todas las columnas en un solo vector llamado 'features'
input_cols = [f"{c}_indexed" for c in nominal_vars] + numeric_vars
assembler = VectorAssembler(inputCols=input_cols, outputCol="features_raw")

# Escalado de variables (recomendado para modelos como Regresión Logística)
scaler = StandardScaler(inputCol="features_raw", outputCol="features")

# Construir y ejecutar el Pipeline de Preparación
pipeline_spark = Pipeline(stages=indexers + [assembler, scaler])
model_preproc = pipeline_spark.fit(df_spark)
df_final = model_preproc.transform(df_spark)

# Dividir en entrenamiento (70%) y prueba (30%)
train_data, test_data = df_final.randomSplit([0.7, 0.3], seed=42)

# Mostrar resultado
print(f"Registros Entrenamiento: {train_data.count()}")
print(f"Registros Prueba: {test_data.count()}")
train_data.select("features", "label").show(5, truncate=False)
```

```python
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator

# Configurar y entrenar el modelo
rf = RandomForestClassifier(labelCol="label", featuresCol="features", numTrees=100, seed=42)
model_rf = rf.fit(train_data)

# Generar predicciones en el conjunto de prueba
predictions = model_rf.transform(test_data)

# Evaluación del modelo
# AUC-ROC
evaluator_bin = BinaryClassificationEvaluator(labelCol="label", rawPredictionCol="rawPrediction", metricName="areaUnderROC")
auc_roc = evaluator_bin.evaluate(predictions)

# Accuracy y F1-Score
evaluator_multi = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")

accuracy = evaluator_multi.evaluate(predictions, {evaluator_multi.metricName: "accuracy"})
f1_score = evaluator_multi.evaluate(predictions, {evaluator_multi.metricName: "f1"})

# Mostrar métricas
print(f"Resultados del Modelo:")
print(f"- ROC-AUC:  {auc_roc:.4f}")
print(f"- Accuracy: {accuracy:.4f}")
print(f"- F1-Score: {f1_score:.4f}")

# Mostrar tabla con label, predicción y probabilidad
predictions.select("label", "prediction", "probability").show(10, truncate=False)
```

```python
import pandas as pd
import numpy as np
from scipy import stats

# Intervalo de Confianza al 95% para la media de puntaje_satisfaccion
data = df['puntaje_satisfaccion']
media = np.mean(data)
sem = stats.sem(data)  # Error estándar de la media
confianza = 0.95

# Usamos la distribución t de Student
intervalo = stats.t.interval(confianza, len(data)-1, loc=media, scale=sem)

print(f"Media de satisfacción: {media:.2f}")
print(f"Intervalo de confianza (95%): ({intervalo[0]:.2f}, {intervalo[1]:.2f})")

# Prueba de Hipótesis (Prueba t de Student para muestras independientes)
# Planteamiento:
# H0 (Hipótesis Nula): La media de satisfacción es IGUAL en ambos grupos (u1 = u2).
# H1 (Hipótesis Alternativa): La media de satisfacción es DIFERENTE (u1 != u2).

grupo_completado = df[df['curso_completado'] == 1]['puntaje_satisfaccion']
grupo_no_completado = df[df['curso_completado'] == 0]['puntaje_satisfaccion']

# Realizar el test t (asumiendo varianzas diferentes con equal_var=False - Test de Welch)
t_stat, p_value = stats.ttest_ind(grupo_completado, grupo_no_completado, equal_var=False)

alpha = 0.05

print("\n--- Resultados de la Prueba de Hipótesis ---")
print(f"Estadístico t: {t_stat:.4f}")
print(f"Valor-p: {p_value:.4e}")

# Conclusión
if p_value < alpha:
    print(f"Conclusión: Se RECHAZA la hipótesis nula (H0).")
    print("Existen evidencias estadísticas suficientes para afirmar que el puntaje de satisfacción es significativamente diferente entre quienes completan el curso y quienes no.")
else:
    print(f"Conclusión: NO se rechaza la hipótesis nula (H0).")
    print("No hay evidencia suficiente para decir que completar el curso afecte el puntaje de satisfacción.")
```
